In Orellana et al. 2016, we have used all the crystal structures of a homomeric protein to create an ensemble. What we call an ensemble in this article was a set of structures that are not chimeric, represent the whole protein rather than a single domain of it, and do not contain more than 5 consecutive non-terminal missing residues. Using this ensemble, we have performed principal component analysis (PCA) to quantify the diversity of the structures and also assign states more objectively. We have also projected the MD trajectories to this space to understand the transitions.

This tutorial provides a workflow for the PCA. There are two steps, first step is creation of the principal component space using an experimental ensemble. If you have prepared your ensemble (with/without pdbParser), you can simply import that pdb file here to perform the PCA. For projections you can select any of the structures. What we have done in that article is outlined below:
1) Align the structures to the closed state structure (this is easier to guess, without through knowledge of an open state)
2) Calculate the PCA in this aligned ensemble (PCA is still calculated with the average structure of the ensemble)
3) Project the structures using the closed state as the reference.

When each structure is aligned to closed state and the reference for projections is the closed state it becomes easier to identify the states and the transitions.

Once you have the ensemble, the second step is the projection of the trajectories. Saving your trajectories is going to be easy if you have kept the original PDB numbering. You can simply use the information extracted during ensemble preparation.
